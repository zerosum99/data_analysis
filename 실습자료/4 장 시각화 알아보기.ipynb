## 함수를 지정해서 내부 칼럼 추가

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

### apply 함수를 이용해서 행의 값들을 변경하고 새로운 열에 추가하기

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
tips.head()
```

## 그래프 그리기

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

### 점 그래프에 대한 함수 기능 조회

```python
help(axes1.scatter)
```

### 마커사이즈를 s에 추가

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 20,    
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

### 칼라를 c에 추가

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 30,    
    c=tips['sex_color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

### 원그래프 그리기

```python
# matplotlib 컬러맵 생성
import matplotlib as mpl
import numpy as np

def make_colors(n, colormap=plt.cm.Spectral):
    return colormap(np.linspace(0.1, 1.0, n))

def make_explode(n):
    explodes = np.zeros(n)
    explodes[0] = 0.15
    return explodes
```

### 한글 폰트 처리

```python
from matplotlib import font_manager, rc

if sys.platform == 'darwin' :
    rc('font', family='AppleGothic')

    plt.rcParams['axes.unicode_minus'] = False
else :
    
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
```

```python
labels= [ "자전거","버스","지하철"]

sizes = [90, 60 , 45]
```

```python

plt.pie(sizes, labels=labels, colors=make_colors(3), explode=make_explode(3), autopct='%1.1f%%', shadow=True, startangle=135)
plt.axis('equal')
plt.show()
```

## 여러 그래프를 연결

```python
iris = sns.load_dataset("iris")
```

```python

joint = sns.jointplot("petal_length", "sepal_length", data=iris,
                          marginal_kws=dict(bins=15, rug=True),
                       annot_kws=dict(stat="r"),
                        s=40, edgecolor="w", linewidth=1)

joint.set_axis_labels(xlabel='petal_length', ylabel="sepal_length")
joint.fig.suptitle('Joint Plot of iris', fontsize=10, y=1.03)
```
